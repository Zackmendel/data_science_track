```python
independent_var = ["biking", "smoking"]
dependent_var = "heart_disease"

X = np.array(df[independent_var])

from statsmodels.stats.outliers_influence import variance_inflation_factor

# Loop through each column to calculate VIF
for i in range(X.shape[1]):
    vif_value = variance_inflation_factor(X,i)
    print(f"Variance Inflation Factor for {df[independent_var].columns[i]}: {vif_value:.2f}")

# If VIF > 10, there is strong multicolinearity and Vice versa
```

#### Findings
From our pair plots, there is a strong linear relationship between heart disease and biking and a weak linear relationship between heart disease and smoking. Smoking and biking do not have a clear linear relationship with plots randomly scattered without forming a clear linear path. All the above deductions were confirmed by our correlation plot with values:
- -0.94 for heart disease vs biking showing a strong negative correlation between them.
- 0.31 for heart disease vs smoking showing a weak correlation between them.
- 0.015 for smoking vs biking show no correlation between them.
- VIF are both less than 5, meaning that they are okay.

With the above findings, we can use a multiple linear regression model to fit our data with heart disease as the predictive variable. Smoking and biking are not correlated to each other, we will be exploring this later in the analysis and we will be checking how this affects our model's efficiency.

# Fitting our Model using Statsmodel Library

```python
formula_str = dependent_var+' ~ '+'+'.join(independent_var)
# formula_str = dependent_var+' ~ '+'+'.join(independent_cols)

model = smf.ols(formula=formula_str, data=df)
fitted = model.fit()

print(fitted.summary())
# fitted.summary()
```

#### Findings:
Lets start by interpreting the regression results.
1. Model Fit Metrics
    - R-Squared(0.980): this is the proprtion of variance in the dependent variable that is explained by our model. It usually ranges fro 0-1 and the closer to 1 the better. In our case the R-squared value is very good.
    - Adjusted R-Squared(0.98): this penalizes for adding irrelevant presictors to the model. The value should be as close to the R-squared value as possible. In our case, we do not have irrelevant variables since our values are the same.
    - F-Statistic(11900) & Prob(FStatistic(0.00)): this tests if our overall regression is statistically significant. The F-stats value should be large and the P(F) should be <0.05. In our case, we satisfy both conditions well meaning our model is statistically significant.
2. Model Diagnostics
    - Log Likelihood(-493.68): this is used for comparing different models. The hegher(less negative) the model is, the better the model. This is not relevant for now since we are not comparing our model to another model.
    - AIC(993.4) & BIC(1006.0): also used for comparing with other models and lower values shows a better fit with lesser predictive features. this is also not relevant for now since we are not comparing models.
3. Coefficients Table:
    -  Coeff: is the estimated effect of each predictor variable on the dependent variable
    -  Std err: standard error of the coefficient(precision of the estimate)
    -  t: is the t-statistic(coeff/std err)
    -  P>|t|: this is the p-value for each predictor. And as usual, a value <0.05 shows that the predictor variable is statistically significant, >0.05 shows weak evidence meaning that predictor MAY not matter when other variables are included. In our case, all the P>|t| values are <0.005 intication that all our predictors are statistically significant.
4. Residual Normality Tests:
    - Omnibus(2.794) & Prob(Omnibus(0.247)): this is used to test if the residuals are normally distributed. p>0.05 indicates that residuals are consistent to a normal distribution. Our values are higher than 0.05so our residuals should be normally distributed.
    - Skew(-0.141): measures the assymetry of residuals. 0 means perfectly symmetric while values -0.5 and 0.5 are considered to be acceptable. Our skew is slightly below 0 and is considered acceptable.
    - Jarque-Bera(2.582) & Prob(JB(0.275)): same interpretation as Omnibus, p>0.05 is considerred okay. This means that our residuals are also consistent with normal distribution.
    - Kurtosis(3.211): this measures the tailedness of residuals, values close to 3 are acceptable. This shows that the tailedness of our residuals is in the normal range.
5. Autocorrelation
    - Durbin-Watson(1.917): this tests the residual autocorrelation. It ranges from 0-4, with values closer to 2 being acceptable. Our value is close to 2 which shows that there is no autocorrelation in our model.
  
6. Multicollinearity:
    - Condition Number(125): This tests for predictors being highly correlated(multicolinearity). Values <30 is safe, <1000 needs to be checked, >1000 shows a high risk of multicollinearity. In our case, the condition number is well within acceptable range indicating low risks of multicolinearity.
  
#### Conclusion
From our findings above, we can conlude that our model is fitting the data well and satisfies all the conditions for a good linear regression fit.

## 📊 Template for Regression Result Interpretation

### 1. Model Fit
- **R-squared**: [value] → proportion of variance explained. Closer to 1 = stronger fit.
- **Adjusted R-squared**: [value] → adjusts for number of predictors. Should be close to R²; a big drop suggests irrelevant predictors.
- **F-statistic & Prob(F-statistic)**: [value, p-value] → tests overall model significance. Large F and p < 0.05 = model is statistically valid.

---

### 2. Coefficients
- **Coefficient (coef)**: Estimated effect of predictor on dependent variable.
- **Standard error (std err)**: Precision of estimate.
- **t-statistic & p-value (P>|t|)**: Tests if predictor is significant.
  - p < 0.05 → predictor is statistically significant.
  - Confidence intervals crossing zero → predictor not significant.
- **Interpretation**: Highlight which predictors matter most and their direction (positive/negative).

---

### 3. Model Diagnostics
- **Log-Likelihood**: Higher (less negative) = better fit. Used for model comparison.
- **AIC & BIC**: Lower values = better fit with fewer predictors. Useful for comparing models.

---

### 4. Residual Normality
- **Omnibus & Jarque-Bera tests**: p > 0.05 → residuals consistent with normal distribution.
- **Skew**: Between -0.5 and +0.5 = acceptable symmetry.
- **Kurtosis**: Close to 3 = normal tailedness.

---

### 5. Autocorrelation
- **Durbin-Watson**: Range 0–4.
  - ~2 = no autocorrelation.
  - <1 or >3 = problematic.

---

### 6. Multicollinearity
- **Condition Number**:
  - <30 = safe.
  - 30–1000 = mild risk.
  - >1000 = severe multicollinearity.
- If high, check Variance Inflation Factors (VIFs) to identify problematic predictors.

---

### 7. Conclusion
- Summarize:
  - Does the model fit well (R², F-test)?
  - Are predictors individually significant?
  - Do residuals meet assumptions (normality, independence, homoscedasticity)?
  - Any multicollinearity concerns?
- End with a clear statement:  
  *“The model explains X% of variance, key predictors are [list], assumptions are largely satisfied, and multicollinearity is [low/mild/high]. Overall, the model is [reliable/limited] for prediction and interpretation.”*


# Check for Indepence: Plot of Predictor Variables vs Model Residuals

```python
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
fig.subplots_adjust(hspace=0.5, wspace=0.2)
fig.suptitle("Check for Independence: Plot of Predictors VS Model Residuals", fontsize=16, y=1.05)
axs.ravel()

for index, column in enumerate(df.columns):
    axs[index].set_title(f"{column}")
    axs[index].grid(True)
    axs[index].scatter(x=df[column], y=fitted.resid, color='blue', edgecolor='k')
    xmin=min(df[column])
    xmax=max(df[column])
    axs[index].hlines(y=0, xmin=xmin*0.7, xmax=xmax*1.1, color='red', linestyle='--', linewidth=3)
    if index==0:
        axs[index].set_ylabel("Residuals")
    

plt.show()
```

#### Findings:

From visualy inspecting our independence plot, we can assertain that there are residuals present and the fact that no distinct pattern or clustters is an indication that the residuals(errors) are independent. The random distribution of the residuals around zero shows that there are no external factors influencing the residuals.

# Check for Homoscedaticity

There are two tests here:
1. Visual Inspection: Plot of Fitted VS Residual values for the model
2. Breush Pagan Test

```python
plt.figure(figsize=(10,5))
plt.scatter(x=fitted.fittedvalues, y=fitted.resid, color='blue', edgecolor='k')
xmax=max(fitted.fittedvalues)
plt.hlines(y=0, xmin=0, xmax=xmax*1.02, color='red', linestyle='--', linewidth=3)
plt.grid(True)
plt.title("Plot of Fitted VS Residual Values")
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")

plt.show()
```

```python
residuals = fitted.resid

bp_test_result = sms.het_breuschpagan(residuals, fitted.model.exog)
print("Breusch Pagan Result:")
print("LM Statistic:", bp_test_result[0])
print("LM-Test p-value:", bp_test_result[1])
print("F-Statistic:", bp_test_result[2])
print("F-Test p-value:", bp_test_result[3])
```

#### Findings:
By visually inspecting the plot of fitted vs. residual values, we can deduce that there is no discernable pattern and our residuals are randomly scattered around zero indicating homoscadaticity.

Furthermore, from our Berush-Pagan test, the p-values of LM-Test and F-test are greater than 0.05(although slightly), there is no enough evidence to reject the null hypothesis of homoscedacity. This means that our model is performing well and the variance of the residuals is canstant accross all predictor variables.

This suggests that the model’s predictions are reliable across the range of fitted values, as the error variance remains stable.

# Check for Normality
There two ways to confirm this:
1. Histogram of Residuals
2. Quartile-Quartile(Q-Q) Plot

```python
plt.figure(figsize=(10,5))
plt.hist(residuals, color='blue', edgecolor='k')
plt.title("Histogram of Residuals", fontsize=16)
plt.xlabel("Residuals")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(10,5))
fig=qqplot(fitted.resid_pearson, fit=True, line='45')
plt.title("Q-Q Plot of Normalised Residuals", fontsize=16)
plt.grid(True)

plt.show()
```

#### Findings:
As we can recall from our regression report earlier, our Ominibus, Jarque Bera, Skew and Kurdosis already confirmed a normal distribution. But further tests to prove this will not be misplaced.
From visual inspection, the residuals follow a a bell like shape indicating a normal distribution.
The Q-Q plot also shows that it is a normal distribution since a large number of plots lies on the 45 degrees line, although there are some outliers along -3 & -2 that moved slighly away from the line. These should be further investigated.
Minor deviations at the tails are common and generally acceptable unless they indicate systematic skewness or heavy tails. In this case, the residuals appear sufficiently normal for reliable inference.”

#### Conclusion:
Overall, both visual and statistical evidence confirm that the residuals are approximately normally distributed, satisfying the normality assumption required for valid hypothesis testing and confidence intervals.

# Check For Outliers

```python
influence = OLSInfluence(fitted)
(c, p) = influence.cooks_distance

# Plot Cook's distance
plt.figure(figsize=(8, 5))
plt.title("Cook's distance plot for the residuals", fontsize=16)
plt.plot(np.arange(len(c)), c, marker='o', linestyle='-')
plt.xlabel('Observation index')
plt.ylabel("Cook's Distance")
plt.grid(True)
plt.show()
```

#### Findings:
A rule of thumb for cook's distance is that the values does not exceed 1 and that it does not exceed 4 divided by the total number of observations in the dataset.
In our case, all values are below 1 and the value of 4 divided by total number of observations(498) is approximately 0.00803. By visually inspecting, there a quite a considerable number of outliers in dataset, although their impact on the model's prediction is not pronounced, they should also be noted. Ideally, the model should be trained again without these values to see how much they affect its predictability.

#### Colclusion:
Since all Cook’s distances are well below the conventional thresholds, no single observation appears to unduly influence the model. The regression results can therefore be considered stable.
