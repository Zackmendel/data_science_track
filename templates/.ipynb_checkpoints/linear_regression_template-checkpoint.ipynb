### 📊 Simple Linear Regression Project Template

**Objective:** To model the relationship between an independent variable ($X$) and a dependent variable ($y$) and evaluate the model's ability to generalize to unseen data.

#### Step 1: Import Necessary Libraries
Start by importing all required packages for data handling, visualization, and machine learning.

```python
# Core Data Manipulation
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt

# Machine Learning & Evaluation Metrics
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

#### Step 2: Import & Explore the Dataset (EDA)
Load your dataset and perform preliminary checks to understand the distribution and cleanliness of the data.

```python
# Load the dataset
# df = pd.read_csv("path_to_your_dataset.csv")

# Example placeholder dataset (so the template runs without a real file):
df = pd.DataFrame({'Independent_Variable_Name': np.random.rand(100) * 10, 'Dependent_Variable_Name': np.random.rand(100) * 10})

# 1. View the first few rows
display(df.head())

# 2. Check statistical summary (mean, std, min, max, quartiles)
display(df.describe().T)

# 3. Check for missing/null values
print("Total Null Values:", df.isnull().sum().sum())
```

*📝 **Analysis Note:** Look at the standard deviation and mean. Ensure there are no null values before proceeding. If there are null values, impute or drop them.*

---

#### Step 3: Check for Linearity (Visual Inspection)
Linear regression requires a linear relationship between the variables. Verify this visually using a scatter plot.

```python
plt.figure(figsize=(10, 5))
plt.scatter(df["Independent_Variable_Name"], df["Dependent_Variable_Name"], color='blue', edgecolor='k')
plt.xlabel("X-axis Label (Independent)")
plt.ylabel("Y-axis Label (Dependent)")
plt.title("Scatter Plot to Check for Linearity", fontsize=16)
plt.show()
```

*📝 **Analysis Note:** Confirm whether the points follow a general straight-line trend (positive or negative).*

---

#### Step 4: Feature Preparation & Initial Model Fitting
Separate your data into $X$ (features) and $y$ (target). For simple linear regression in Scikit-Learn, ensure $X$ is a 2D array.

```python
X = df["Independent_Variable_Name"].values
y = df["Dependent_Variable_Name"].values

# Reshape X because sklearn expects a 2D array for features
X_reshaped = X.reshape(-1, 1)

# Initialize and fit the model on the full dataset (Baseline)
model = LinearRegression()
model.fit(X_reshaped, y)

# Generate predictions
y_pred = model.predict(X_reshaped)
```

#### Step 5: Visualize the Regression Line (Full Data)
Plot the line of best fit over your original data points.

```python
plt.figure(figsize=(10, 5))
plt.scatter(X, y, color='blue', edgecolor='k', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.title("Linear Regression Fit")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.legend()
plt.show()
```

#### Step 6: Initial Model Evaluation
Extract the equation of the line and check how well the model fits the data using standard error metrics.

```python
print("--- Baseline Model Metrics ---")
print("Slope (Coefficient):", round(model.coef_[0], 4))
print("Intercept:", round(model.intercept_, 4))
print("RSS (Residual Sum of Squares):", round(sum((y - y_pred)**2), 4))
print("R² Score:", round(metrics.r2_score(y, y_pred), 4))
print("MAE (Mean Absolute Error):", round(metrics.mean_absolute_error(y, y_pred), 4))
print("MSE (Mean Squared Error):", round(metrics.mean_squared_error(y, y_pred), 4))
print("RMSE (Root Mean Squared Error):", round(np.sqrt(metrics.mean_squared_error(y, y_pred)), 4))
```

#### Step 7: Train-Test Split for Generalization
To prove the model isn't just memorizing data (overfitting), split it into training and testing sets.

```python
# Split the data (e.g., 80% Train, 20% Test)
X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y, test_size=0.2, random_state=40)

# Initialize and fit the NEW model only on training data
model_tt = LinearRegression()
model_tt.fit(X_train, y_train)

# Generate predictions for both Train and Test sets
y_pred_train = model_tt.predict(X_train)
y_pred_test = model_tt.predict(X_test)
```

#### Step 8: Visualize the Train/Test Split Results
Create plots to visually distinguish how the model fits the training data vs. the unseen test data.

```python
# Combined Plot
plt.figure(figsize=(10, 5))
plt.scatter(X_train, y_train, color='blue', edgecolor='k', label='Training Data')
plt.scatter(X_test, y_test, color='green', edgecolor='k', label='Test Data')
plt.plot(X_test, y_pred_test, color='red', linewidth=2, label='Regression Line')
plt.title("Combined Linear Regression Plot (Train vs Test)")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.legend()
plt.show()
```

#### Step 9: Final Model Evaluation & Comparison
Compare the metrics between the Training set and the Test set.

```python
print("--- Training Set Metrics ---")
print("R² Score Train:", round(metrics.r2_score(y_train, y_pred_train), 4))
print("RMSE Train:", round(np.sqrt(metrics.mean_squared_error(y_train, y_pred_train)), 4))

print("\n--- Test Set Metrics ---")
print("R² Score Test:", round(metrics.r2_score(y_test, y_pred_test), 4))
print("RMSE Test:", round(np.sqrt(metrics.mean_squared_error(y_test, y_pred_test)), 4))
```

*📝 **Analysis Note:** If Test $R^2$ is roughly equal to or slightly higher than Train $R^2$, your model generalizes well. If Test $R^2$ is significantly lower, the model is overfitting.*
