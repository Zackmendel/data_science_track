## Numpy Array Attributes

```python
np.random.seed(1)

ds_1 = np.random.randint(5, size=8)
ds_2 = np.random.randint(5, size=(2, 8))
ds_3 = np.random.randint(5, size=(2, 4, 8))

print("First Dataset:", ds_1) 
print("Second Dataset:", ds_2)
print("Third Dataset:", ds_3)
```

```python
# Numpy array has the attributes: shape, number of dimensions, dtype, size, item size and total size.

print("Dimensions of Third Dataset:", ds_3.ndim, "dimensions")
print("Shape of Third Dataset:", ds_3.shape)
print("Size of Third Dataset:", ds_3.size)
print("Data type of Third Dataset:", ds_3.dtype)
print("Item Size of Third Dataset:", ds_3.itemsize, "bytes")
print("Total Size of Third Dataset:", ds_3.nbytes, "bytes")
```

## Array Indexing

```python
# 1. Indexing is similar to python's indexing, put number in square brackets startinging from 0

# Using our first dataset(one-dimensional)
print("First Dataset:", ds_1)
print("First Element:", ds_1[0])
print("Third Element:", ds_1[2])

# You can access the elements from the pther end using negative index
print("The last Element:", ds_1[-1])
print("The Third Element from the Last:", ds_1[2])
```

```python
# 2. You can access a multidimensional element using a comma seperated tuple of indices

# Using our second dataset(two-dimensional)
print("Second Dataset:", ds_2)
print("First Element:", ds_2[0, 0])
print("Third Element:", ds_2[0, 2])
print("Fourth Element on the Second Row:", ds_2[1, 3])

# You can access the elements from the pther end using negative index
print("The last Element on the Second Row:", ds_2[1, -1])
print("The Third Element from the Last on the First Row:", ds_2[0, 2])
```

## Array Slicing

```python
# This follows the syntax: df[start:stop:step]

# For easy assimilation, lets use an ordered list here
df = np.arange(11)
df
```

```python
# Examples using one-dimensional subarrays

print("----------------------------------------------------------------------------------")
print("First Five(5) Elements df[:5] =", df[:5])
print("All Elements after index 5 df[5:]", df[5:])
print("Some elements in the middle df[3:7]", df[3:7])
print("All even elements df[::2]", df[::2])
print("All stepped element starting from index 1 df[1::2]", df[1::2])

# Reversing the array by using negative steps
print("----------------------------------------------------------------------------------")
print("👈Reversed Array df[::-1]:", df[::-1])
print("👈Stepped Reverse from 7 df[7::-2]:", df[7::-2])
```

```python
# Examples using two-dimensional subarrays

print("----------------------------------------------------------------------------------")
print("Second Dataset =", "\n", ds_2)
print("----------------------------------------------------------------------------------")
print("First Five(5) Elements of the two rows ds_2[:5, :5] =", "\n", ds_2[:5, :5])
print("All Elements after index 5 for two rows ds_2[:2, 5:] =", "\n", ds_2[:2, 5:])
print("Some elements in the middle ds_2[:2, 3:7] =", "\n", ds_2[:, 3:7])
print("All stepped elements by 2 ds_2[:, ::2] =", "\n", ds_2[:, ::2])
print("All stepped element starting from index 1 on both rows ds_2[:, 1::2] =", "\n", ds_2[:, 1::2])

# Reversing the array by using negative steps
print("----------------------------------------------------------------------------------")
print("👈Reversed Array for all rows ds_2[:2, ::-1]:", "\n", ds_2[:2, ::-1])
print("👈Stepped Reverse from 7 for all rows ds_2[:, 7::-2]:\n", ds_2[:, 7::-2])
```

```python
# Clone our ds_2
main_ds = ds_2.copy()
```

```python
# Subarrays as no-copy views
print("Second Dataset =", "\n", main_ds)
```

```python
# Create a 3 by 2 array of our dataset
sub_array = main_ds[:2, :3]
print("Subarray =\n", sub_array)

# Make a change to the subarray
sub_array[0,0] = 10
print("First value of subarray is changed to 10 =\n", sub_array)

# View the original dataset
print("Original dataset is now =\n", main_ds)
```

```python
# Use .copy() to create a 3 by 2 array of our dataset
sub_array = ds_2[:2, :3].copy()

print("Subarray =\n", sub_array)

# Make a change to the subarray
sub_array[0,0] = 5
print("First value of subarray is changed to 10 =\n", sub_array)

# View the original dataset
print("Original dataset is now =\n", ds_2)
```

#### Array Concatenation

```python
# You can use the np.concatenate, np.vstack and np.hstack for joining two or more arrays together.
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
z = np.array([7, 8, 9])

print("Concatenation: np.concatenate([x, y, z]) = ", np.concatenate([x, y, z]))
print("Vstack: np.vstack([x, y, z]) =\n", np.vstack([x, y, z]))
print("Hstack: np.hstack([x, y, z]) =\n", np.hstack([x, y, z]))
```

#### Array Splitting

```python
# You use the np.split, np.hsplit and np.vsplit for spliting arrays into smaller subaarays.
x = np.arange(10)
x_1, x_2, x_3, x_4 = np.split(x, [3, 5, 7])
print(x_1, x_2, x_3, x_4)
# It is important to note that the [3, 5, 7] used above splits the data from those points, that is, @3: x[:3], @5: x[3:5], x[5:7] and @7: x[7:].
```

```python
x = np.arange(16).reshape(4, 4)
x_1, x_2 = np.hsplit(x, 2)
print("Horrizontal Split(hsplit):", "\nx_1 =", x_1, "\nx_2 =", x_2)
```

```python
x_1, x_2 = np.vsplit(x, 2)
print("Vertical Split(vsplit):", "\nx_1 =", x_1, "\nx_2 =", x_2)
```
